```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df= pd.read_csv('file/train.csv') 
```

```python
df.head(10)
```

- PassengerId: Unique identifier for each passenger.
- Survived: 1 if the passenger survived, 0 if they did not.
- Pclass: Ticket class (1st, 2nd, or 3rd class).
- Name: Passenger's name.
- Sex: Passenger's gender (male or female).
- Age: Passenger's age.
- SibSp: Number of siblings/spouses aboard.
- Parch: Number of parents/children aboard.
- Ticket: Ticket number.
- Fare: Fare paid for the ticket.
- Cabin: Cabin number.
- Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

```python
# Проверим, есть ли отсутствующие значения (NaN) в каждом столбце
missing_data = df.isnull().sum() 

# Выведем столбцы с отсутствующими данными и их количество пропущенных значений
for column, count in missing_data.items():
    if count > 0:
        print(f"{column}: {count} пропущенных значений")

# Подсчитаем общее количество отсутствующих значений в наборе данных
total_missing_values = missing_data.sum()
print(f"Общее количество отсутствующих значений: {total_missing_values}") 
```

```python
Age_count = df['Age'].isnull().sum() 
print(f"Общее количество отсутствующих значений в Age: {Age_count}") 
```

```python
# Вычислим средний возраст пассажиров с использованием NumPy
mean_age = np.mean(df['Age']) 
print(f'Средний возраст пассажиров: {mean_age:.2f} лет') 
print(df['Age'].mean())
```

```python
pclass_xt=pd.crosstab(df['Survived'],
                     df['Age'])  
pclass_xt 
```

```python
# Вычислим средней возраст пассажира каждого класса 
average_age_by_class = df.groupby('Pclass')['Age'].mean() 
average_age_by_class
```

```python
print(average_age_by_class)
```

```python
# Создадим гистограмму
#Это позволяет нам увидеть, как распределен возраст пассажиров
plt.hist(df['Age'].dropna(), bins=20, edgecolor='blue')
plt.xlabel('Возраст')
plt.ylabel('Частота') 
plt.title('Распределение возраста пассажиров')
plt.show() 
```

```python
# Подсчитаем количество мужчин и женщин
gender_counts = df['Sex'].value_counts()

# Создадим круговую диаграмму
#Это может помочь наглядно показать соотношение полов на борту
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Соотношение мужчин и женщин на борту')
plt.show()
```

```python
# Этот график поможет понять, как класс билета влиял на выживаемость
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.xlabel('Класс билета')
plt.ylabel('Число пассажиров')
plt.title('Выживаемость в зависимости от класса билета')
plt.legend(title='Survived', labels=['Не выжил', 'Выжил'])
plt.show() 
```

```python
# Этот график поможет нам понять, как порт посадки влиял на выживаемость.
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.xlabel('Порт посадки')
plt.ylabel('Число пассажиров')
plt.title('Выживаемость в зависимости от порта посадки')
plt.legend(title='Survived', labels=['Не выжил', 'Выжил'])
plt.show() 
```

```python
# Создадим график плотности распределения
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Выжил', shade=True)
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Не выжил', shade=True)
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.title('Плотность распределения возраста с разделением на выживших и не выживших')
plt.legend()
plt.show() 
```
