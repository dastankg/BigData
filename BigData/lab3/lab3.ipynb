```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi/2, np.pi/2, 1000)  # Создаем массив значений x от -pi/2 до pi/2
y = np.tan(x)  # Вычисляем значения y = tan(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('График функции y = tan(x) на интервале (-pi/2, pi/2)')
plt.ylim(-10, 10)  # Установим ограничение по оси y для наглядности
plt.show()

```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)  # Создаем массив значений x от -2*pi до 2*pi
y = np.sin(x)  # Вычисляем значения y = sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('График функции y = sin(x)')
plt.ylim(-1.5, 1.5)  # Установим ограничение по оси y для наглядности
plt.show()
```

```python
# Создание двух subplot
fig, (ax1, ax2) = plt.subplots(1, 2)

# График 1 - гармошка
x =  np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.sin(x)
ax1.plot(x, y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Гармошка')

# График 2 - парабола
x = np.arange(-5, 5, 0.1)
y = x ** 2 - 2 * x + 1
ax2.plot(x, y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Парабола')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

# График 1 - гармошка
x1 = x =  np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y1 = y = np.sin(x)
ax.plot(x1, y1, label='Гармошка')

# Увеличиваем масштаб графика "гармошка" по оси y
ax.set_ylim(-0.5, 1.1)

# График 2 - парабола
x2 = np.arange(-5, 5, 0.1)
y2 = x2 ** 2 - 2 * x2 + 1
ax.plot(x2, y2, label='Парабола')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графики на одной оси')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
R = 2.0

# Параметр t (угол), который будет меняться от 0 до 2*pi
t = np.linspace(0, 2 * np.pi, 100)

# Вычисляем x(t) и y(t) для каждого значения t
x = R * np.cos(t)
y = R * np.sin(t)

# Создаем график параметрической функции
plt.figure(figsize=(6, 6))  # Размер графика
plt.plot(x, y, label='Окружность', color='blue')  # Построение графика
plt.xlabel('x')  # Метка по оси x
plt.ylabel('y')  # Метка по оси y
plt.title('Параметрическая функция: Окружность')  # Заголовок графика
plt.legend()  # Добавление легенды
plt.grid(True)  # Включение сетки

# Отображение графика
plt.show()
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
data = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.pie(data, labels=years, autopct='%.1f%%')
plt.title('Динамика прожиточного минимума в КР')

plt.show()
```

```python

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
data = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

# Создаем 3D-график
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Создаем scatter plot с маркерами
ax.scatter(years, data, [0] * len(years), c=data, cmap='viridis', s=100)

# Настройка меток и заголовка
ax.set_xlabel('Год')
ax.set_ylabel('Прожиточный минимум, сом')
ax.set_zlabel('Значение')
ax.set_title('Зависимость Значения от Года и Прожиточного минимума')

# Добавляем цветовую шкалу
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap='viridis'), ax=ax)
cbar.set_label('Значение')

# Отображаем график
plt.show()
```
