```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

breast = load_breast_cancer()
```

```python
df = pd.DataFrame(breast.data, columns=breast.feature_names)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast.data , breast.target, random_state=0)
```

```python
feature_names = breast.feature_names[:6]
X_train = X_train[:, :6]

df = pd.DataFrame(X_train, columns=feature_names)

# Ð£Ð¼ÐµÐ½ÑŒÑˆÐ°ÐµÐ¼ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð², Ð±ÐµÑ€ÐµÐ¼ Ñ‚Ð¾Ð»ÑŒÐºÐ¾ Ð¿ÐµÑ€Ð²Ñ‹Ðµ 4 Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°
scatter_matrix = pd.plotting.scatter_matrix(df.iloc[:, :4], c=y_train,
                                            figsize=(10, 10), marker='o',
                                            hist_kwds={'bins': 20},
                                            s=80, alpha=.8)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```
