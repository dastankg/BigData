```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
breast = load_breast_cancer()
```

```python
df = pd.DataFrame(breast.data, columns=breast.feature_names)
df["target"] = breast.target
```

```python
df
```

```python
df["target"].value_counts()
```

```python
corr_matrix = df.corr()
fig, ax = plt.subplots(figsize=(10, 7))
ax = sns.heatmap(corr_matrix)
```

```python
X = breast.data[:]
y = breast.target[:]

# Split the data using Scikit-Learn‘s train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.20,
                                                    random_state=0)
```

```python
best = 0
c = 0
for i in range(2, 11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    n = np.mean(y_pred == y_test)
    if n > best:
        best = n
        c = i
print(f'Лучший результат = {best}\nКолво соседей = {c}')
```

<h3>

В этом коде используется датасет breast cancer из библиотеки scikit-learn.

Сначала данные загружаются и преобразуются в dataframe, чтобы удобнее было работать с признаками и целевой переменной.

Далее выводится описательная статистика по целевой переменной и строится корреляционная матрица признаков.

Затем данные разбиваются на обучающую и тестовую выборки при помощи train_test_split.

После этого запускается цикл перебора гиперпараметра количества соседей для модели K Nearest Neighbors от 2 до 10. Для каждого значения строится модель, делается предсказание на тестовой выборке и подсчитывается доля верных предсказаний. Лучшее значение и соответствующее количество соседей запоминаются.

В конце выводятся оптимальное значение метрики (доли верных предсказаний) и лучшее количество соседей.

Таким образом, в работе подбирается оптимальное количество соседей для алгоритма KNN на датасете рака молочной железы. Это позволяет построить модель с наилучшим качеством предсказаний.
